```python
import pandas as pd
from sklearn.linear_model import LinearRegression
```

```python
# Cargar los datos de Kaggle de Happiness

# Definir la ruta donde se encuentra la carpeta y los CSVs que iremos trabajar
ruta_datos_preprocesados = '../data/ficheros_procesados/datos_preprocesados_limpios.csv'

# Creación del DataFrame
df = pd.read_csv(ruta_datos_preprocesados, delimiter=",", engine="python")
df
```

```python

# Preparar los datos
X = df[['Year', 'Health (Life Expectancy)']]
y = df['Happiness Score']

#y = df['Happiness_Rank']

# Crear el modelo de regresión lineal
model = LinearRegression()

# Entrenar el modelo
model.fit(X, y)

# Predecir el ranking de expectativa de vida en 10 años
future_life_expectancy = [[2033, 80]]  # Valores hipotéticos para 10 años en el futuro
future_happiness_rank = model.predict(future_life_expectancy)

# Imprimir la predicción
print("El ranking de expectativa de vida en 10 años será:", future_happiness_rank)
```
