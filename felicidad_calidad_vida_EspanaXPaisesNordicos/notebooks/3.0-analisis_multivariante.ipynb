## Análisis multivariante

Gráficos para analizar más de dos variables al mismo tiempo.

* sns.boxplot
* sns.pairplot()
* sns.replot() con hue, col, row, size
* sns.heatmap()
* sns.clustermap()

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('../data/ficheros_procesados/datos_preprocesados_limpios.csv')
df.head(10)
```

```python
# sns.catplot permite dibujar varios gráficos categóricos en una figura
# Gráfico de conteo para mostrar la distribución de colores en cada categoría de corte cut
sns.catplot(data=df, x='Freedom', col='Country', kind='count')
```

```python
sns.catplot(data=df, x='Country', y='Generosity', col='Year', kind='box')
```

```python
sns.catplot(data=df, x='Generosity', y='Country', col='Year', kind='bar', errorbar=None)
```

```python
sns.catplot(data=df, x='Country', y='Year', col='Trust (Government Corruption)', kind='point', errorbar=None)
```

```python
plt.figure(figsize=(15, 5))
sns.boxplot(data=df, x='Year', y='Trust (Government Corruption)', hue='Country', palette='viridis')
plt.legend(loc='lower right', bbox_to_anchor=(1, 1))
```

```python
# sns.heatmap
df_corr = df.corr(numeric_only=True)
sns.heatmap(df_corr, annot=True, cmap='viridis');
```

```python
import numpy as np

df_corr = df.corr(numeric_only=True)
filter = np.triu(df_corr) # np.tril para el triángulo superior
sns.heatmap(df_corr, annot=True, cmap='viridis', mask=filter);
```

```python
import matplotlib.pyplot as plt
import pandas as pd

plt.figure(figsize=(20,15), dpi=300)
df = pd.get_dummies(df)
df_corr = df.corr()
filter = np.triu(df_corr) 
sns.heatmap(df_corr, annot=True, cmap='viridis', mask=filter, fmt=".2f");
```

```python
# correlación con barras para la columna Generosity

df_corr = df.corr() # matriz de correlación
corr_price = df_corr['Generosity'] # filtrar correlación Generosity
corr_price = corr_price.sort_values(ascending=False)[1:] # ordenar descendente
plt.figure(figsize=(12, 5))
sns.barplot(x=corr_price.index, y=corr_price.values)
plt.xticks(rotation=45)
```

```python
df_corr = df.corr() # matriz de correlación

corr_Social_support = df_corr['Social support'] # filtrar correlación Social support
filter = (corr_Social_support < -0.1) | (corr_Social_support > 0.1) # quitar aquellas con baja correlación
corr_Social_support = corr_Social_support[filter]
corr_Social_support = corr_Social_support.sort_values(ascending=False)[1:] # ordenar descendente


plt.figure(figsize=(12, 5))
ax = sns.barplot(x=corr_Social_support.index, y=corr_Social_support.values)
plt.xticks(rotation=45)

labels = [f'{value:.3f}' for value in corr_Social_support.values] # redondear las correlaciones a dos decimales
# labels = [str(round(value, 2)) for value in corr_Social support.values]
ax.bar_label(ax.containers[0], labels=labels);
```
